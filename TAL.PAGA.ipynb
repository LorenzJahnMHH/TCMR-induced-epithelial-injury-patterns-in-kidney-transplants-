```python
import scanpy as sc
import numpy as np
import random

random.seed(0)
```

```python
import os

os.chdir(r'D:\R_Ordner\KTx Daten')
```

```python
adata_tal = sc.read_h5ad("D:/R_Ordner/KTx Daten/Ktx Objekte final/Maus/tal_mouse.h5ad")

healthy_tal_clusters = ['cTAL1', 'cTAL2', 'mTAL1', 'mTAL2', 'cmTAL1', 'cmTAL2', 'MD']

adata_tal.obs['largersubct'] = adata_tal.obs['celltype_to_plot'].apply(
    lambda x: 'healthyTAL' if x in healthy_tal_clusters else x
)
```

```python
sc.pp.highly_variable_genes(adata_tal, flavor="seurat", n_top_genes=500)

sc.tl.pca(adata_tal, mask_var="highly_variable")
sc.pp.neighbors(adata_tal, n_pcs=30)
sc.tl.diffmap(adata_tal)
sc.tl.paga(adata_tal, groups='largersubct')
adata_tal.uns['iroot'] = np.flatnonzero(adata_tal.obs['largersubct'] == 'healthyTAL')[0]
sc.tl.dpt(adata_tal)


```

```python
sc.set_figure_params(figsize=(11, 11), dpi=540)
sc.pl.paga(adata_tal, 
           color='dpt_pseudotime', 
           node_size_scale=24, 
           edge_width_scale=2, 
           threshold=0.15, 
           cmap='viridis',
           frameon=False,  # Verhindert die Anzeige des Plots
           labels=None,  # Entfernt die Subcluster-Namen
           colorbar=False) 
```
